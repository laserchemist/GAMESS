#Chemistry Periodic Trends
Encoded in the periodic table are myriad atomic properties arranged to highlight trends in properties. We can visualize these properties in the traditional view of the periodic table. Each element is distinguished by a fixed number of protons unique to that element such as 9 for fluorine (F). This number of protons is also known as the atomic number, z. A neutral atom has the same number of electrons as protons or 9 as in fluorine. Fluorine has a single isotope with 10 netrons and when adding protons and neutrons we get 19 also labelled as ${^{19}F}$. In chlorine two majors isotopes which occur naturally at 75.5% ${^{35}Cl}$ and 23.9% ${^{37}Cl}$ as shown below. The rows of the periodic table are based on the electron configuration. The first row corresponds to filling the 1s atomic orbital with helium having the maximum of 2 electrons in a stable configuration known as ${1s^2}$. Neon has both the 2s and the 2p orbitals full in a stable ${1s^22s^22p^6}$ configuration. Below these configurations are shown to require the most energy to remove an electron or ionize.
